```python
import pandas as pd
df = pd.read_excel('Road_Accident_Fatalities.xlsx')
display(df.head())
print(df.shape)
```

```python
#1
import numpy as np
df['Helmet Used'] = df['Helmet Used'].replace(np.nan, 'Unknown')
df['Helmet Used'] = df['Helmet Used'].replace('N/A', 'Unknown')
print(df['Helmet Used'].value_counts())
```

```python
#2
import pandas as pd

# Convert 'Date' column to datetime objects
try:
    df['Date'] = pd.to_datetime(df['Date'])
except ValueError as e:
    print(f"Error converting 'Date' column to datetime: {e}")
    # Handle the error appropriately, e.g., skip the conversion or log the error

# Extract the month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Ensure 'Fatalities' and 'Injuries' columns are numeric
for col in ['Fatalities', 'Injuries']:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    except ValueError as e:
        print(f"Error converting '{col}' column to numeric: {e}")

display(df.head())
```

```python
#3
# Group the data by month and count accidents
monthly_accidents = df.groupby('Month')['Accident ID'].count()

# Group the data by weather condition and count accidents
weather_accidents = df.groupby('Weather Condition')['Accident ID'].count()

# Display the results
print("Monthly Accident Counts:\n", monthly_accidents)
print("\nWeather Condition Accident Counts:\n", weather_accidents)
```

```python
#4
import matplotlib.pyplot as plt
import seaborn as sns

# Line chart for monthly accidents
plt.figure(figsize=(10, 6))
plt.plot(monthly_accidents.index, monthly_accidents.values, marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.title('Monthly Trend of Road Accidents')
plt.xticks(monthly_accidents.index)  # Ensure all month labels are displayed
plt.grid(True)
plt.savefig('monthly_accident_trend.png')
plt.show()

# Heatmap for weather conditions and accident counts
plt.figure(figsize=(10, 6))
sns.heatmap(pd.DataFrame(weather_accidents).rename(columns={'Accident ID':'Accident Count'}), annot=True, cmap="YlGnBu", fmt='g')
plt.xlabel('Weather Condition')
plt.ylabel('Accident Count')
plt.title('Impact of Weather Conditions on Accident Counts')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('weather_accident_heatmap.png')
plt.show()
```

```python
#5
# Calculate fatality rate for each road type
df['Fatality Rate'] = df.groupby('Road Type')['Fatalities'].transform(lambda x: x / x.sum())

# Ensure 'Road Type' is categorical
df['Road Type'] = df['Road Type'].astype('category')

# Group data by 'Road Type' and calculate fatality rate
df_fatality_rates = df.groupby('Road Type')['Fatality Rate'].mean().reset_index()

# Create a DataFrame for speeding-related accidents
df_speeding = df[df['Speeding'] == 'Yes']

# Calculate fatality rate for speeding accidents by road type
df_speeding['Fatality Rate'] = df_speeding.groupby('Road Type')['Fatalities'].transform(lambda x: x / x.sum())

# Group data by 'Road Type' and calculate fatality rate for speeding accidents
df_speeding_fatality_rates = df_speeding.groupby('Road Type')['Fatality Rate'].mean().reset_index()

display(df_fatality_rates)
display(df_speeding_fatality_rates)
```

```python
#6
# Convert 'Emergency Response Time (mins)' to numeric, coercing errors to NaN
df['Emergency Response Time (mins)'] = pd.to_numeric(df['Emergency Response Time (mins)'], errors='coerce')

# Handle missing values by filling with the median
median_response_time = df['Emergency Response Time (mins)'].median()
df['Emergency Response Time (mins)'].fillna(median_response_time, inplace=True)

# Verify data quality (check for inconsistencies like negative values)
print(df['Emergency Response Time (mins)'].describe())
```
